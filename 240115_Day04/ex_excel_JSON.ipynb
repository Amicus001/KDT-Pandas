## [ JSON -> DataFrame 저장]

(1). 데이터 준비<hr>

```python
file3 = '../DATA/nobel.json'
```

(2) Json -> DataFrame

```python
nobelDF = pd.read_json(file3)
nobelDF.info()
```

(3) 데이터 확인

```python
nobelDF.head(3), nobelDF.tail(3)
```

```python
#행 추출: iloc()/loc()
nobelDF.iloc[0]
```

```python
nobelDF.iloc[0]['prizes']
```

```python
# 행 추출 = iloc[[n1, ...,]]/loc[]
nobelDF.iloc[[0,3]]
```

```python
oneData = nobelDF.iloc[0]['prizes']
type(oneData)
```

```python
oneData.keys(), oneData.items()
```

```python
for k,v in oneData.items():
    print(f'[{k}]\n-->{v}')
    if k == 'laureates' : 
        for d in v:
            for k1,v1 in d.items():
                print('-', k1,v1)
```

- Json 중첩데이터의 다중 반복문 ==> 단일 리스트 형태로 처리 : 평탄회(정규화)

```python
pd.json_normalize(oneData)
```

```python
nobelDF2 = pd.json_normalize(oneData, record_path = 'laureates', meta = ['year', 'category'] )
nobelDF2.info()
```

```python
pd.json_normalize(oneData, record_path = 'laureates', meta = ['year', 'category'])
```

- 모든 행에 대한 정규화(평탄화) 진행

컬럼 읽기: DF 객체변수명['컬럼명'], DF객체변수명.컬럼명
- > 모든 행에 record_path에 해당하는 key가 존재해야 함!
- > 없는 행에 record_path에 해당하는 key 추가하기

```python
pd.json_normalize(nobelDF.prizes)
for data in nobelDF.prizes:
    print(type(data),data.keys(), sep= '\n\n')
    #해당 dict에 laureates key가 없다면 추가하기.
    data.setdefault('laureates', [])
```

(7) Json의 다양한 방식의 DataFrame 저장: orient parameter

```python
# data = key:{value}
json_data='''
[
{"id" : "1","name":"sravan","age":22},
{"id" : "2","name":"harsha","age":22},
{"id" : "3","name":"deepika","age":21},
{"id" : "4","name":"jyothika","age":23}
]
'''
```

- orient = 'record' 설정: json 파일 데이터를 행 단위로 읽어서 DataFrame에 저장함.

```python
df= pd.read_json(json_data, orient = 'records')
print (df)
```

```python
json_data2 = '''
{
    "student-1": {"id" : "1", "name":"sravan", "age":22},
    "student-2":  {"id" : "2", "name":"harsha", "age":22}, 
    "student-3":  {"id" : "3","name":"deepika","age":21},
    "student-4":  {"id" : "4","name":"jyothika","age":23}
    }'''
```

```python
df = pd.read_json(json_data2, orient = 'index')
print (df)
```

```python
df = pd. read_json(json_data2).T #orient = 'column' <default
print(df)
```

```python
#json format

json_data3 = '''
[
    [ "1", "sravan",22 ],
    [ "2", "harsha",22 ],
    [ "3", "deepika",21 ],
    [ "4", "jyothika",23 ]
]
'''
```

```python
df = pd.read_json(json_data3,) #orient = 'values') #값으로만 구성된 테이블
print(df)
```
