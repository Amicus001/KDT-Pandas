(1) 최적화하기

```python
netflix['type'] = netflix['type'].astype('category')
```

(2) 제목이 "Limitless"인 모든 행 찾기

```python
netflix[netflix['title'] == 'Limitless']
```

(3) Robert Rodriguez 감독이 제작한 'Movie'유형의 모든 콘텐츠 찾기

```python
directorisrobert = (netflix['director'] == 'Robert Rodriguez')
itismovie =  ( netflix ['type'] == 'Movie')
netflix[directorisrobert & itismovie]
```

(4) 등록된 날짜가 2019-07-31 이거나 감독이 Robert Altman인 모든 콘텐츠 찾기

```python
isdate = (netflix['date_added'] == '31-Jul-19')
isAltman = (netflix['director'] == 'Robert Altman')
netflix[isdate | isAltman]
```

(5) 감독이 Orson Welles, Aditya Kripalani 또는 Sam raimi인 모든 콘텐츠 찾기

```python
isorson = (netflix['director'] == 'Orson Welles')
isAditya = (netflix['director'] == 'Aditya Kripalani')
isSam = (netflix['director'] == 'Sam Raimi')

netflix[isorson|isAditya|isSam]
```

(6)2019/5/11 ~ 2019/6/11 사이에 등록된 모든 콘텐츠 찾기

```python
date = (netflix['date_added'] <= '11-May-19')
date2 = (netflix['date_added']>= '11-Jun-19')
netflix[date & date2]
```

(7) director 열에서 NaN 값이 있는 모든 행 삭제하기

```python
netflix.dropna(subset = ['director']) #★
```

(8) 넷플릭스가 콘텐츠를 단 하나만 등록한 날짜 식별하기

```python
netflix.drop_duplicates(subset = ['date_added'], keep = False)
```
