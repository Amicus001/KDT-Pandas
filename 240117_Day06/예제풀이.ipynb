## 240117. 과제: matplotlib 실습, 3장. 

1.Module Loading <hr>

```python
import numpy as np
import matplotlib.pyplot as plt
```

2.Load Data<hr>

```python
x = np.linspace(-4,4,1024)
y = .25 * (x + 4.) * (x + 1) * (x - 2)
```

3.GraphMaking <hr>

```python
plt.title('A polynomial') #그래프 제목
plt.plot(x,y, c = 'k')
plt.show() #출력하기
```

4.LaTeX style <hr>

```python
x2 = np.linspace(-4,4,1024)
y2 = .25 * (x2+4.) * (x2+1.) * (x2-2)

plt.title('$(x2)\\frac{1}{4}(x2+4)(x2+1)(x2-2)$')
plt.plot(x2,y2,c = 'k')
plt.show()
```

5.Add Labels<hr>

```python
X = np.linspace(-4,4,1024)
Y = .25 * (X+4.) * (X+1) * (X-2)

plt.title('Power Curve for airfoil KV873') #제목
plt.xlabel('Air Speed') #X축 라벨
plt.ylabel('Total drag') #Y축 라벨
plt.plot(X, Y, c= 'k') #
plt.show()
```

6.Add Text<hr>

```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X+4) * (X+1) *(X-2)

plt.text(-0.5, -0.25, 'Brackmard minimum') #x축 -0.5지점, y축 -0.25.지점에 'Brackmard minimum' 텍스트 출력할게
plt.plot(X,Y, c='k')
plt.show
```

7. Border Box Setting<hr>

```python
X = np.linspace(-4, 4, 1024)
Y = .25 * (X+4) * (X+1) *(X-2)
box = {'facecolor':'.75', 'edgecolor':'k', 'boxstyle':'round'}
plt.text(-0.5, -0.25, 'Brackmard minimum', bbox = box)
plt.plot(X,Y, c='k')
plt.show
```

9. Legends<hr>

```python
X = np.linspace(0,6,1024)
Y1 = np.sin(X)
Y2 = np.cos(X)

plt.xlabel('X')
plt.ylabel('Y')

plt.plot(X,Y1,c = 'k', lw = 3., label = 'sin(X)')
plt.plot(X,Y2,c = '.5', lw = 3., ls = '--', label = 'cos(X)')

plt.legend() #범례 렌더링, 각 선의 라벨명으로.
plt.show()
```

10. Add Grid<hr>

```python
X = np.linspace(-4,4,1024)
Y = .25 * (X+4.)*(X+1.)*(X-2.)

plt.plot(X,Y,c='k')
plt.grid(True, lw=2, ls = '--', c='.75')
plt.show()
```

11. Grid gap control<hr> 

```python
import matplotlib.ticker as ticker

x = np.linspace(-15,15,1024)
y = np.sinc(x)

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))

plt.plot(X, Y, c = 'k')
plt.show()
```

11. Adding Marking Lable control <hr>

```python
import numpy as np
import random

name_list = ('Omar', 'Serguey', "Max", 'Zhou', "Abidin")
value_list = np.random.randint(0,99,size = len(name_list))
pos_list = np.arange(len(name_list))

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.FixedLocator((pos_list)))
ax.xaxis.set_major_locator(ticker.FixedFormatter((name_list)))

plt.bar(pos_list, value_list, color = '.75', align= 'center' )
plt.show()
```
