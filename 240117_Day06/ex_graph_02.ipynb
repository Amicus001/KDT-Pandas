```python
import matplotlib.pyplot as plt  #그래프 관련 모듈
import pandas as pd  #matplotlib 패키지
import matplotlib.font_manager as fm #패키지 리소스(자원) 중 폰트 관련 모
import matplotlib as mpl
```

- 데이터 준비

```python
#2D 그래프를 그리기 위한 데이터 준비
x = range(1, 100, 10)
y = [d**2 for d in x]
```

- 그래프 그리기
 - plot(data)= y축 데이터.
 - plot(data1, data2)= x축에 data1, y축에 data2 사용
 - 반환값: 라인/선 그래프가 객체 리스트에 담겨서 반환됨.

```python
ret = plt.plot(x,y) #반환값 = line2D
type(ret[0]), len(ret)
```

### X,Y축 레이블 및 눈금 설정
 - x,y측 데이터 설명-> plt.xlabel(),plt.ylabel()
 - 눈금 간격/출력 형태 설정-> plt.xticks, plt.yticks(), plt.tick_param()

```python
# label

#그래프 그리기
plotlist=plt.plot(y,'ro--',y2,'g^', label = '걍... 그거')
plt.title('[테스트]')

#x축 y축 레이블 지정하기
plt.xlabel('value', loc = 'right', fontsize='xx-large', style = 'italic')
plt.ylabel(' result ', loc = 'top', mouseover = True )

#눈금 ticks
plt.xticks(ticks = [0,2,4,6,8] , labels = ['A', 'B', 'C', 'D', 'E'] )
plt.yticks(color = 'blue')

# 범례 설정
plt.legend(["인절미절임", "인절미꿀절임"],loc = 'center')

#그래프 안에 텍스트 출력하기
plt.text(2,2000,"강사님언제쉬어요")
```

### 텍스트 추가
- 그래프 내에 위치 지정 후 텍스트 추가 설정: plt.text(x시작좌표, y시작좌표, 출력할텍스트)

```python
# label

#그래프 그리기
plotlist=plt.plot(y,'ro--',y2,'g^', label = '걍... 그거')
plt.title('[테스트]')

#x축 y축 레이블 지정하기
plt.xlabel('value', loc = 'right', fontsize='xx-large', style = 'italic')
plt.ylabel(' result ', loc = 'top', mouseover = True )

#눈금 ticks
plt.xticks(ticks = [0,2,4,6,8] , labels = ['A', 'B', 'C', 'D', 'E'] )
plt.yticks(color = 'blue')

# 범례 설정
plt.legend(["인절미절임", "인절미꿀절임"],loc = 'center')

#그래프 안에 텍스트 출력하기
box = {'facecolor': '.75', 'edgecolor' :'k', 'boxstyle' : 'round', 'alpha':0.5} 
plt.text(2,4000,"강사님언제쉬어요", backgroundcolor='yellow', color='blue', bbox = box)
```

### 화살표 설정하기
- 그래프 특정 지점에 화살표 표기하기
- text와 함께 종종 사용됨.
- plt.annotate()

```python
# label

#그래프 그리기
plotlist=plt.plot(y,'ro--',y2,'g^', label = '걍... 그거')
plt.title('[테스트]')

#x축 y축 레이블 지정하기
plt.xlabel('value', loc = 'right', fontsize='xx-large', style = 'italic')
plt.ylabel(' result ', loc = 'top', mouseover = True )

#눈금 ticks
plt.xticks(ticks = [0,2,4,6,8] , labels = ['A', 'B', 'C', 'D', 'E'] )
plt.yticks(color = 'blue')

# 범례 설정
plt.legend(["인절미절임", "인절미꿀절임"],loc = 'center')

#그래프 안에 텍스트 출력하기
box = {'facecolor': '.75', 'edgecolor' :'k', 'boxstyle' : 'round', 'alpha':0.5} 
plt.text(2,4000,"강사님언제쉬어요", backgroundcolor='yellow', color='blue', bbox = box)

# 화살표 설정하기
plt.annotate()
```
