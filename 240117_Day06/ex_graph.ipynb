# [데이터 시각화 맛보기]

- 모듈 로딩

```python
import matplotlib.pyplot as plt
import pandas as pd
```

- 데이터 준비

```python
#2D 그래프를 그리기 위한 데이터 준비
x = range(1, 100, 10)
y = [d**2 for d in x]
```

- 그래프 그리기
 - plot(data)= y축 데이터.
 - plot(data1, data2)= x축에 data1, y축에 data2 사용
 - 반환값: 라인/선 그래프가 객체 리스트에 담겨서 반환됨.

```python
ret = plt.plot(x,y) #반환값 = line2D
type(ret[0]), len(ret)
```

## - 기본: 라인(선)그래프

- 시간의 변화를 표현하는 그래프
- 연속적이거나 순서가 있는 데이터의 시각화에 사용

- 라인(선)그래프에서 형식(포맷)지정: str타입의 포맷 설정하기.
  * 색상, 형태, 선 스타일: 'ro--' : red cirlle dot-line
  * 많이 사용되는 색상: r - red, g - green, b - blue, k - black
  * 많이 사용되는 형태: o - circle, ^ - pyramid, < - left-pointed triangle, > - right-pointed triangle, v - down-pointed triangle
  * 많이 사용되는 스타일: - : 실선, --: 점선, -. : dashdot

```python
ret2 = plt.plot(x,y ,'gv-')
```
