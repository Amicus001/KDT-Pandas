## [실습] 다양한 정렬

```python
import numpy as np
from math import nan

data = {'col1' : ['A', 'A', 'B', nan, 'D', 'C'],
        'col2' : [2, 1, 9, 8, 7, 4],
        'col3' : [0, 1, 9, 4, 2, 3],
        'col4' : ['a', 'B', 'c', 'D', 'e', 'F']}
        
```

```python
dataDF = pd. DataFrame(data)
dataDF
```

- col1 column 기준으로 오름차순

```python
dataDF.sort_values(by='col1')
#NaN: Not a Number;(결측치) 값의 위치- 기본적으로 가장 마지막
```

```python
#Nan (빈칸데이터, 결측치)값의 위치 설정-> na_position parameter
dataDF.sort_values(by='col1',na_position = 'first')
```

```python
#col4 컬럼을 기준으로 오름차순 정렬
dataDF.sort_values(by = 'col4', key= lambda col:col.str.upper())
```
