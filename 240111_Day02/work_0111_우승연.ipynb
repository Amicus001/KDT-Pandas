(6)Column별 데이터 속성(index, values, shape, ndim,types)

```python
for colname in movieDF.columns:
    print(f"index : {movieDF[colname].index}")
    print(f"vaules : {movieDF[colname].values}")
    print(f"shape : {movieDF[colname].shape}")
    print(f"ndim : {movieDF[colname].ndim}")
    print(f"size : {movieDF[colname].size}")
    print(f"dtype : {movieDF[colname].dtype}")
```
